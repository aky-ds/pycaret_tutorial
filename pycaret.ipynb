<h1>Introduction to PyCaret</h1>
PyCaret is an open-source, low-code machine learning library in Python that automates workflows.
It simplifies the process of training and deploying machine learning models with minimal code.

Key Features:
- Automates ML model training and selection
- Provides easy-to-use APIs for classification, regression, and clustering
- Supports deployment and model monitoring

```python
# importing the dataset and libraries
import pandas as pd
from pycaret.datasets import get_data
data = get_data('juice')
from pycaret.classification import *
```

```python
data['Store7'].value_counts()
```

```python
# Setting up the dataset for data modelling
s=setup(data,target='Store7',session_id=134)
```

```python
models=compare_models()
```

```python
# Plot the models
plot_model(models,plot='auc')
```

```python
# Evaluating the model using confusion matrix
plot_model(models, plot = 'confusion_matrix')
```
